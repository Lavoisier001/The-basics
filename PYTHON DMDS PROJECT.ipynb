```python

users = {}
next_id = 1



def register():
    global next_id

    email = input("Enter your email address: ")
    name = input("Enter your full name: ")
    password = input("Choose a password: ")

    
    if "@" not in email:
        return {"status": "error", "message": "Invalid email format"}

    
    if len(password) < 8 or not any(char.isdigit() for char in password):
        return {"status": "error", "message": "Password must be at least 8 characters long and contain at least one digit"}

    
    for user in users.values():
        if user["email"] == email:
            return {"status": "error", "message": "User email already taken"}

    
    user_id = next_id
    users[user_id] = {
        "email": email,
        "name": name,
        "password": password,
        "is_admin": False
    }
    next_id += 1

    return {"status": "success", "message": "User registered successfully", "user_id": user_id}



def login(email: str, password: str) -> dict:
    for user_id, user in users.items():
        if user["email"] == email and user["password"] == password:
            return {"status": "success", "message": "Login successful", "user_id": user_id}
    return {"status": "error", "message": "Invalid email or password"}



def remove_user(user_id: int) -> dict:
    if user_id in users:
        del users[user_id]
        return {"status": "success", "message": "User removed successfully"}
    else:
        return {"status": "error", "message": "User not found"}



def list_users() -> dict:
    if not users:
        return {"status": "error", "message": "No users found"}

    user_list = []
    for user_id, user in users.items():
        user_list.append({
            "id": user_id,
            "email": user["email"],
            "name": user["name"]
        })

    return {"status": "success", "users": user_list}



def update_password(user_id: int, old_password: str, new_password: str) -> dict:
    
    if user_id not in users:
        return {"status": "error", "message": "User not found"}

    user = users[user_id]

    
    if user["password"] != old_password:
        return {"status": "error", "message": "Old password is incorrect"}

    
    if len(new_password) < 8 or not any(char.isdigit() for char in new_password):
        return {"status": "error", "message": "New password must be at least 8 characters long and contain at least one digit"}

    
    user["password"] = new_password
    return {"status": "success", "message": "Password updated successfully"}



print("---- REGISTER ----")
result = register()
print(result)
print(users)

print("\n---- LIST USERS ----")
print(list_users())

print("\n---- UPDATE PASSWORD ----")
try:
    user_id = int(input("Enter user ID to update password: "))
    old_pass = input("Enter old password: ")
    new_pass = input("Enter new password: ")
    print(update_password(user_id, old_pass, new_pass))
except ValueError:
    print("Invalid ID entered!")

print("\n---- LIST USERS AGAIN ----")
print(list_users())


 
```
